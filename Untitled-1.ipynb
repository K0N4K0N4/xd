```python
import math

def es_primo(n: int) -> bool:
    if n>0:
        if n == 1 or n == 2:
            return True
        if n%2 == 0:
            return False
        for i in range(3, math.floor(math.sqrt(n))+1):
            if n%i == 0:
                return False 
        else:
            return True
    return False

n = int(input("Ingresa tu número"))
analisis = es_primo(n)

if analisis is True:
    print(n, " es primo.")
else:
    print(n, " no es primo.")
```

```python
def lista(l: list, k: int) -> list:
    n = len(l)
    k_norm = k % n
    inicio = l[-k_norm: ]
    final = l[ :-k_norm]
    return inicio + final

l = [1, 2, 3, 4]
k = 2
resultado = lista(l, k)
print(resultado)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def euler(f, a, b, y0, n):
    delta_t = (b - a) / n
    T = np.array([])
    Y = np.array([])
    T = np.append(T, a)
    Y = np.append(Y, y0)
    for i in range(1, n + 1):
        T = np.append(T, a + i * delta_t)
        Y = np.append(Y, Y[i-1] + f(T[i-1], Y[i-1]) * delta_t)
    return Y

def f(t, x):
    return r * x

a = 0
b = 5
r = 0.5
y0 = 10



# Euler N=50
n = 50
Y50 = euler(f, a, b, y0, n)
T50 = np.linspace(a, b, n + 1)

# Euler N=200
n = 200
Y200 = euler(f, a, b, y0, n)
T200 = np.linspace(a, b, n + 1)

# Solución analítica
T_exacta = np.linspace(a, b, 1000)
X_exacta = y0 * np.exp(r * T_exacta)

# Gráficas
plt.plot(T_exacta, X_exacta, label='Solución analítica')
plt.plot(T50, Y50, '--', label='Euler N=50')
plt.plot(T200, Y200, '--', label='Euler N=200')

plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.grid(True)
plt.show()
```
