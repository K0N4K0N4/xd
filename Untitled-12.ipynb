```python
import math

def es_primo(n: int) -> bool:
    if n>0:
        if n == 1 or n == 2:
            return True
        if n%2 == 0:
            return False
        for i in range(3, math.floor(math.sqrt(n))+1):
            if n%i == 0:
                return False 
        else:
            return True
    return False

n = int(input("Ingresa tu número"))
analisis = es_primo(n)

if analisis is True:
    print(n, " es primo.")
else:
    print(n, " no es primo.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def euler(f, a, b, y0, n):
    delta_t = (b - a) / n
    T = np.array([])
    Y = np.array([])
    T = np.append(T, a)
    Y = np.append(Y, y0)
    for i in range(1, n + 1):
        T = np.append(T, a + i * delta_t)
        Y = np.append(Y, Y[i-1] + f(T[i-1], Y[i-1]) * delta_t)
    return Y

def f(t, x):
    return r * x

a = 0
b = 5
r = 0.5
y0 = 10



# Euler N=50
n = 50
Y50 = euler(f, a, b, y0, n)
T50 = np.linspace(a, b, n + 1)

# Euler N=200
n = 200
Y200 = euler(f, a, b, y0, n)
T200 = np.linspace(a, b, n + 1)

# Solución analítica
T_exacta = np.linspace(a, b, 1000)
X_exacta = y0 * np.exp(r * T_exacta)

# Gráficas
plt.plot(T_exacta, X_exacta, label='Solución analítica')
plt.plot(T50, Y50, '--', label='Euler N=50')
plt.plot(T200, Y200, '--', label='Euler N=200')

plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_modificado(f, g, a, b, x0, y0, n):
    delta_t = (b - a) / n
    T = np.array([])
    Y = np.array([])
    X = np.array([])
    T = np.append(T, a)
    X = np.append(X, x0)
    Y = np.append(Y, y0)
    for i in range(1, n + 1):
        T = np.append(T, a + i * delta_t)
        Fx = f(T[i-1], X[i-1], Y[i-1])
        Gy = g(T[i-1], X[i-1], Y[i-1])
        X = np.append(X, X[i-1] + Fx * delta_t)
        Y = np.append(Y, Y[i-1] + Gy * delta_t)
    return X, Y

def f(t, x, y):
    return x * (3 - x - 2 * y)

def g(t, x, y):
    return y * (2 - x - y)

ganan_conejos = [(0.1, 0.3), (2.5, 1.9), (2.6, 1.4)]
ganan_ovejas = [(0.1, 0.1), (1.2, 1.8), (2.6, 2)]
ptos_equilibrio = [(0, 0), (3, 0), (0, 2), (1, 1)]
n = 4000
a = 0
b = 20
for ex, ey in ptos_equilibrio:
    plt.plot(ex, ey, 'o', label=f'Pto. eq. ({ex}, {ey})')
for x, y in ganan_conejos + ganan_ovejas:
    X1, Y1 = euler_modificado(f, g, a, b, x, y, n)
    plt.plot(x, y, 'x', label=f' ({x}, {y})')
    plt.plot(X1, Y1, '-')

plt.xlabel('conejos')
plt.ylabel('ovejas')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_atras(f, delta_t, M, z0):
    Z = np.empty((M + 1, 2))
    Z[0] = z0
    for i in range(1, M + 1):
        z = Z[i-1] - f(Z[i-1]) * delta_t
        Z[i] = z
    return Z

def f(z):
    x, y = z
    return np.array([x * (3 - x - 2 * y), y * (2 - x - y)])

p= np.array([1.0, 1.0])
v_s = np.array([2.0, np.sqrt(2.0)])
epsilon = 0.05
z0_plus = p + epsilon * v_s / np.linalg.norm(v_s)
z0_minus = p - epsilon * v_s / np.linalg.norm(v_s)
delta_t = 0.005
M = 6000


Z1= euler_atras(f, delta_t, M, z0_plus)
Z2 = euler_atras(f, delta_t, M, z0_minus)
plt.plot(Z1[:, 0], Z1[:, 1], label='Rama estable')
plt.plot(Z2[:, 0], Z2[:, 1], label='Rama inestable')


xg = np.linspace(0.0, 3.2, 21)
yg = np.linspace(0.0, 2.2, 21)
X, Y = np.meshgrid(xg, yg)
U = X * (3 - X - 2 * Y)
V = Y * (2 - X - Y)
plt.quiver(X, Y, U, V, angles='xy')
plt.xlim(0, 3.2)
plt.ylim(0, 2.2)
plt.legend()
plt.grid(True)
plt.show()
```
